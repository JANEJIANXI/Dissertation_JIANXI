```python
import numpy as np
import pandas as pd
import math

def normalize(x):
    min_v = -0.02777778
    max_v = 0.7777778
    return (x - min_v)/(max_v-min_v)

cutoff = 0.6

datas1 = pd.read_excel('tanimoto_matrix_GS.xlsx')
datas2 = pd.read_excel('functional_similarity_matrix_GS.xlsx')
cols = list(datas1)

result=[]

for i in range(0,datas1.shape[0]):
    for j in range(i+2,datas1.shape[0]+1):
        if datas1.iloc[i,j] >= cutoff:
            result.append(datas1.iloc[i,0]+'\t'+cols[j]+'\t'+str(datas1.iloc[i,j])+'\t'+str(normalize(datas2.iloc[i,j])))

print(result)
```

```python
# save the matrix

df = pd.DataFrame(result)
COF = pd.ExcelWriter('Find_out_candidates_via_cutoff.xlsx')
df.to_excel(COF)
COF.save()
print('end')
```
